Common Input features for models: There are 4 models -
1. GBM - Gradient boosting 
2. ANN - Artificial Neural Network
3. RNN - Recurrent Neural Network
4. LGBM - Light Gradient Boosting 
There are 18 columns common for all 4 algorithms. Most of them are columns in our dataframe, but there are some columns which are not in our df and is derived from the 
existing columns. My understanding and details of these columns:
1. transaction_log_id - The unique identifier to identify the transaction. There are no missing value. In this dataframe, this is a unique number and the length of transaction_log_id = length of df.  (Only used in GBM - In my opinion, we can't use this to create a model. We can use this for evaluation purpose)
2. merchant_id - Unique Id represent a merchant. No missing values. There are 1029 unique values in our dataframe
3. location_id - Unique id to represent a location. No missing values, there are 137886 unique locations in our df.
4. product_id - Identifier to identify the product. No missing values. 7043 unique products
5. terminal_number - Identify the terminal which is used. No missing values, 436733 unique terminal ids. Used in GBM, ANN and RNN
6. action_name - No missing values, 32 unique values. We may have to do One-Hot-Encoding 
7. amount - No missing values, Mean value: 287.955, Max value: 500000.0 and Min value: 0
8. adjusted_amount - No missing values, Mean: 265.56, Max: 500000.0, min: -469900.0
9. upc - Universal product code of the item. This is like an id. There are no missing items and 7043 unique values in our df. This is exactly the same unique values of product_id. So upc and product_id might be the same. We need to check this two thorughly to understand that. 
10. category_id - This represents the category of the product. We have some missing values in this. There are 10 unique values in this dataframe. Most frequent value is 7. We can add that value to the missing fields. We may need to do One-Hot-Encoding for this too. 
11. multipack_parent - Is the card a multipack parent(I think that shows if the card belongs to a multi pack card category  like 4x25). No missing values. This has 2 unique values - True or False. So we may need to encode(binary encoding) the values. 
12. ignore_multipack_transaction - This shows if the card belongs to a multi-pack or not. There are no missing values, only 2 options - True or False. We can impute values using simple imputer and drop_first=True
13. pos_entry_type - Entry type of point of sale, eg: manual. There are 11 unique items in this dataset and 63% data missing. (This is an ordinal data and the most frequent value is 22. (not sure what is the significance of 22). So we can fill the columns with the mode value.)
In my opinion, we shouldn't use this value for our model building because when more than 20-25% missing, the data may not have standard distribution and this will affect our modeling.

The transaction_date column includes date and time of transaction. In this df, all columns should have same date value since we are analyzing one particular day transaction. But have different value for the time. This may give more idea. We can extract details from that particular column like -
